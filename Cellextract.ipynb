```python
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import re
from scipy.ndimage import rotate
```

```python
def disp_img(img):
    plt.figure(figsize=(6,4))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

margin = 10
```

```python
img = cv2.imread('/content/WhatsApp Image 2022-07-06 at 5.03.14 PM.jpeg') # BGR format
disp_img(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
disp_img(img_gray)
```

```python
img_blur = cv2.GaussianBlur(img_gray, (3, 3), 0)
disp_img(img_blur)
```

```python
img_thresh = cv2.threshold(img_blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
disp_img(img_thresh)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
img_opening = cv2.morphologyEx(img_thresh, cv2.MORPH_OPEN, kernel, iterations=1)

disp_img(img_opening)
```

```python
img_invert = 255 - img_opening
disp_img(img_invert)
```
