```python
df = pd.read_csv("../input/studemploy1/Stud_Employ.csv")
df
```

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
predictors = df.iloc[:, 0:4]
target = df.iloc[:,4]
predictors_train, predictors_test, target_train, target_test=train_test_split(predictors, target,test_size=0.3, random_state = 123)
```

```python
clf=DecisionTreeClassifier()
clf=clf.fit(predictors_train,target_train)
```

```python
clf.predict(predictors_test)
```

```python
from sklearn.tree import plot_tree
ig= plt.figure(figsize=(25,20))
plot_tree(clf, filled=True)
plt.show
```
